```python
import cv2,imageio
import matplotlib.pyplot as plt
import numpy as np
```

```python
image1 = cv2.imread("/kaggle/working/iris-images/iris_images/iris17_24/right/017R_1.png")
image2 = cv2.imread("/kaggle/working/iris-images/iris_images/iris17_24/right/017R_2.png")
image3=cv2.imread("/kaggle/working/iris-images/iris_images/iris17_24/left/017L_1.png")

height = min(image1.shape[0], image2.shape[0])
width = min(image1.shape[1], image2.shape[1])
image1 = cv2.resize(image1, (width, height))
image2 = cv2.resize(image2, (width, height))
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(image1)
plt.title("Image 01")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image2)
plt.title("Image 02")
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(image3)
plt.title("Image 03")
plt.axis('off')

plt.show()

```

```python
grayimg1=cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
grayimg2=cv2.cvtColor(image2,cv2.COLOR_BGR2GRAY)
grayimg3=cv2.cvtColor(image3,cv2.COLOR_BGR2GRAY)
imageio.imwrite("grayimage1.png",grayimg1)
imageio.imwrite("grayimage2.png",grayimg2)
imageio.imwrite("grayimage3.png",grayimg3)
sift1=plt.imread("grayimage1.png")
sift2=plt.imread("grayimage2.png")
sift3=plt.imread("grayimage3.png")
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(sift1)
plt.title("gray image 1")
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(sift2)
plt.title("gray image 2")
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(sift3)
plt.title("gray image 3")
plt.axis("off")
plt.show()
```

```python
histimage1=cv2.calcHist([grayimg1],[0],None,[256],[0,256])
histimage2=cv2.calcHist([grayimg2],[0],None,[256],[0,256])
histimage3=cv2.calcHist([grayimg3],[0],None,[256],[0,256])
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.bar(np.arange(256), histimage1.squeeze(), color='b')
plt.title('Histogramme for l\'image 1')
plt.xlabel('Intensity per pixel')
plt.ylabel('No of pixel')
plt.subplot(1, 3, 2)
plt.bar(np.arange(256), histimage2.squeeze(), color='r')
plt.title('Histogramme for l\'image 2')
plt.xlabel('Intensity per pixel')
plt.ylabel('No dof pixels')

plt.subplot(1, 3, 3)
plt.bar(np.arange(256), histimage3.squeeze(), color='r')
plt.title('Histogramme for l\'image 2')
plt.xlabel('Intensity per pixel')
plt.ylabel('No dof pixels')

plt.tight_layout()
plt.show()
```

```python
grayimg1=cv2.resize(grayimg1,(500,500))
grayimg2=cv2.resize(grayimg2,(500,500))
grayimg3=cv2.resize(grayimg3,(500,500))
imageio.imwrite('grayimageresized1.png',grayimg1)
imageio.imwrite('grayimageresized2.png',grayimg2)
imageio.imwrite('grayimageresized3.png',grayimg3)
sift1=plt.imread('grayimageresized1.png')
sift2=plt.imread('grayimageresized2.png')
sift3=plt.imread('grayimageresized3.png')
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(sift1)
plt.title('gray image 1')
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(sift2)
plt.title('gray image 2')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(sift3)
plt.title('gray image 3')
plt.axis('off')
plt.show()
```

```python
grayimg1=cv2.GaussianBlur(grayimg1,(9,9),1)
grayimg2=cv2.GaussianBlur(grayimg2,(9,9),1)
grayimg3=cv2.GaussianBlur(grayimg3,(9,9),1)
imageio.imwrite("image1guass.png",grayimg1)
imageio.imwrite("image2guass.png",grayimg2)
imageio.imwrite("image3guass.png",grayimg2)
sift1=imageio.imread("image1guass.png")
sift2=imageio.imread("image2guass.png")
sift3=imageio.imread("image3guass.png")
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(sift1)
plt.title('guassian img 1')
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(sift2)
plt.title('guassian img 2')
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(sift3)
plt.title('guassian img 3')
plt.axis("off")
plt.show()
```

```python
grayimg1=cv2.equalizeHist(grayimg1)
grayimg2=cv2.equalizeHist(grayimg2)
grayimg3=cv2.equalizeHist(grayimg3)
imageio.imwrite("grayequalize1.png",grayimg1)
imageio.imwrite("grayequalize2.png",grayimg2)
imageio.imwrite("grayequalize3.png",grayimg3)
detsift1=imageio.imread("grayequalize1.png")
detsift2=imageio.imread("grayequalize2.png")
detsift3=imageio.imread("grayequalize3.png")
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(detsift1)
plt.title("eqializehist1")
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(detsift2)
plt.title("eqializehist2")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(detsift3)
plt.title("eqializehist3")
plt.axis("off")
plt.show()

```

```python
kernel=np.ones((5,5),np.uint8)
grayimg1=cv2.morphologyEx(grayimg1,cv2.MORPH_OPEN,kernel)
grayimg2=cv2.morphologyEx(grayimg2,cv2.MORPH_OPEN,kernel)
grayimg3=cv2.morphologyEx(grayimg3,cv2.MORPH_OPEN,kernel)
imageio.imwrite("morphologyimg1.png",grayimg1)
imageio.imwrite("morphologyimg2.png",grayimg2)
imageio.imwrite("morphologyimg3.png",grayimg2)
detsift1=imageio.imread("morphologyimg1.png")
detsift2=imageio.imread("morphologyimg2.png")
detsift3=imageio.imread("morphologyimg3.png")
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(detsift1)
plt.title("morphology Ex1")
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(detsift2)
plt.title("morphology Ex2")
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(detsift3)
plt.title("morphology Ex3")
plt.axis("off")
plt.show()
```

```python
sift=cv2.SIFT_create()
```

```python
keypoint1,descriptors1=sift.detectAndCompute(grayimg1,None)
keypoint2,descriptors2=sift.detectAndCompute(grayimg2,None)
keypoint3,descriptors3=sift.detectAndCompute(grayimg3,None)
```

```python
imgkeypoint1=cv2.drawKeypoints(grayimg1,keypoint1,None)
imgkeypoint2=cv2.drawKeypoints(grayimg2,keypoint2,None)
imgkeypoint3=cv2.drawKeypoints(grayimg3,keypoint3,None)
```

```python
imageio.imwrite("detectedsift1.png",imgkeypoint1)
imageio.imwrite("detectedsift2.png",imgkeypoint2)
imageio.imwrite("detectedsift3.png",imgkeypoint3)
detsift1=imageio.imread("detectedsift1.png")
detsift2=imageio.imread("detectedsift2.png")
detsift3=imageio.imread("detectedsift3.png")
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(detsift1)
plt.title("keypoint det1")
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(detsift2)
plt.title("keypoint det2")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(detsift3)
plt.title("keypoint det2")
plt.axis("off")
plt.show()
```

```python
bf=cv2.BFMatcher(cv2.NORM_L2,crossCheck=True)
matches=bf.match(descriptors1,descriptors2)
matches=sorted(matches,key=lambda match:match.distance)
matched_image=cv2.drawMatches(image1,keypoint1,image2,keypoint2,matches[:100],None)
matching_rate=len(matches)/min(len(keypoint1),len(keypoint2))*100
print(f"Matching rate: {matching_rate:.2f}%")
if matching_rate < 50:
    print("Images do not match")
else:
    print("Images match")
print("Images of person1: Left eye with diff angles")
scale_percent = 50
width = int(matched_image.shape[1] * scale_percent / 100)
height = int(matched_image.shape[0] * scale_percent / 100)
dim = (width, height)
resized_image = cv2.resize(matched_image, dim, interpolation=cv2.INTER_AREA)
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Disable axis
plt.text(width // 2, height - 20, f"Matching Rate: {matching_rate:.2f}%", color='white', fontsize=7, fontweight='bold', ha='center')
plt.show()
```
