```python
img=np.ones((256,256),np.uint8)
black=np.zeros((128,128),np.uint8)
white=255*np.ones((128,128),np.uint8)
gray1=100*np.ones((128,128),np.uint8)
gray2=200*np.ones((128,128),np.uint8)
img[0:128,0:128]=black
img[0:128,128:256]=white
img[128:256,0:128]=gray1
img[128:256,128:256]=gray2
print("displaying image")
cv2_imshow(img)
cv2.imwrite('/gdrive/My Drive/Colab Notebooks/out.jpg',img)
```

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
img=cv2.imread('/gdrive/My Drive/Colab Notebooks/images.jpg')
print(img.shape)
cv2_imshow(img)
gray1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(gray1.shape)
cv2_imshow(gray1)
ret,thresh1=cv2.threshold(gray1,100,255,cv2.THRESH_BINARY)
print(thresh1.shape)
cv2_imshow(thresh1)
ret,thresh1=cv2.threshold(gray1,100,255,cv2.THRESH_BINARY_INV)
print(thresh1.shape)
cv2_imshow(thresh1)


```
